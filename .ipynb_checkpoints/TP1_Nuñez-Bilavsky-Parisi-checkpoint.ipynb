## Cálculo de la matriz de distancias

Ahora construimos la matriz de distancias entre todos los museos. Como la tierra es un [geoide](https://es.wikipedia.org/wiki/Geoide) (es decir que no es [plana](https://es.wikipedia.org/wiki/Terraplanismo)), el cálculo de distancias no es una operación obvia. Una opción es proyectar a un [sistema de coordenadas local](https://geopandas.org/en/stable/docs/user_guide/projections.html), de forma tal que las distancias euclideas se correspondan con las distancias en metros. En este notebook usamos [EPSG](https://en.wikipedia.org/wiki/EPSG_Geodetic_Parameter_Dataset) 22184. 

```python
# En esta línea:
# Tomamos museos, lo convertimos al sistema de coordenadas de interés, extraemos su geometría (los puntos del mapa), 
# calculamos sus distancias a los otros puntos de df, redondeamos (obteniendo distancia en metros), y lo convertimos a un array 2D de numpy
D = museos.to_crs("EPSG:22184").geometry.apply(lambda g: museos.to_crs("EPSG:22184").distance(g)).round().to_numpy()
```

### Matriz de adyacencia: construimos una matriz conectando a cada museo con los $m$ más cercanos

```python
def construye_adyacencia(D,m): 
    # Función que construye la matriz de adyacencia del grafo de museos
    # D matriz de distancias, m cantidad de links por nodo
    # Retorna la matriz de adyacencia como un numpy.
    D = D.copy()
    l = [] # Lista para guardar las filas
    for fila in D: # recorriendo las filas, anexamos vectores lógicos
        l.append(fila<=fila[np.argsort(fila)[m]] ) # En realidad, elegimos todos los nodos que estén a una distancia menor o igual a la del m-esimo más cercano
    A = np.asarray(l).astype(int) # Convertimos a entero
    np.fill_diagonal(A,0) # Borramos diagonal para eliminar autolinks
    return(A)

m = 3 # Cantidad de links por nodo
A = construye_adyacencia(D,m)
```

## Construcción de la red en NetworkX (sólo para las visualizaciones)

```python
G = nx.from_numpy_array(A) # Construimos la red a partir de la matriz de adyacencia
# Construimos un layout a partir de las coordenadas geográficas
G_layout = {i:v for i,v in enumerate(zip(museos.to_crs("EPSG:22184").get_coordinates()['x'],museos.to_crs("EPSG:22184").get_coordinates()['y']))}
```

```python
fig, ax = plt.subplots(figsize=(15, 15)) # Visualización de la red en el mapa
barrios.to_crs("EPSG:22184").boundary.plot(color='gray',ax=ax) # Graficamos Los barrios
nx.draw_networkx(G,G_layout,ax=ax) # Graficamos los museos
```

## Punto 1:

Partiendo de:
$$
\mathbf{p} = (1-\alpha)\, C\, \mathbf{p} + \frac{\alpha}{N} \mathbf{1} \Rightarrow 
$$

Paso restando a 
$
((1-\alpha)\, C\, \mathbf{p})
$

$$
\mathbf{p} - (1-\alpha)\, C\, \mathbf{p} = \frac{\alpha}{N} \mathbf{1} \Rightarrow
$$

Saco factor comun p:

$$
\left(I - (1-\alpha)\, C\right)\mathbf{p} = \frac{\alpha}{N} \mathbf{1} \Rightarrow
$$

Divido por 
$
\frac{\alpha}{N} :
$

$$
\frac{N}{\alpha} \left(I - (1-\alpha)\, C\right)\mathbf{p} = \mathbf{1} \Rightarrow
$$

Como 
$
M = \frac{N}{\alpha} \left(I - (1-\alpha)\, C\right) y \quad \mathbf{b} = \mathbf{1}
$
$$
\frac{N}{\alpha} \left(I - (1-\alpha)\, C\right)\mathbf{p} = \mathbf{1} \leftrightarrow Mp = b
$$

Partiendo de la ecuacion 3 queda demostrando que el vector de rankings p es la solucion del sistema 
Mp=b

```python
def calcula_matriz_C(A): 
    # Función para calcular la matriz de trancisiones C
    # A: Matriz de adyacencia
    # Retorna la matriz C
    k = A.sum(axis=1)  # suma por filas: grado de cada página
    Kinv = np.diag([1/ki if ki != 0 else 0 for ki in k])
    C = A.T @ Kinv
    return C

def calcular_matriz_M(A, alfa):
    C = calcula_matriz_C(A)
    N = A.shape[0]
    I = np.eye(C.shape[0])
    M = (N / alfa) * (I - (1 - alfa) * C)
    return M

```

## Punto 2:


La condicion que se debe cumplir para que exista una unica solución en esta ecuacion es que  M sea inversible.

Como:
$$
M = \frac{N}{\alpha}(I - (1-\alpha)C)
$$
basta con demostrar que:
$$
I - (1-\alpha)C
$$
es **inversible**.


Definimos:

$$
B = (1-\alpha) C
$$

Queremos demostrar que **( I - B )** es inversible.

Para que esto suceda debemos ver que (demostracion por absurdo mas abajo):

$$
\|B\|_1 < 1
$$

es decir:


$$
\|B\|_1 = (1-\alpha)\|C\|_1 < 1
$$

A partir de esto deducimos que para que M sea inversible debe pasar que :
$$
\|C\|_1 = 1
$$

y

$$ 0 < \alpha < 1 $$

En este caso sabemos que C es una matriz estocástica por columnas, es decir:

$$
\|C\|_1 = \max_j \sum_i |c_{ij}| = 1
$$

Entonces, si $$ 0 < \alpha < 1 $$ se cumple:


$$
\|B\|_1 = (1-\alpha)\|C\|_1 = 1-\alpha < 1
$$


---

**Demostración por el absurdo:**

Supongamos que **( I - B )** no es inversible.

Entonces, existe un vector no nulo $\mathbf{x}$ tal que:

$$
(I - B) \mathbf{x} = \mathbf{0}
$$

lo que implica:

$$
B \mathbf{x} = \mathbf{x}
$$

Tomamos norma 1 en ambos lados:

$$
\|B \mathbf{x}\|_1 = \|\mathbf{x}\|_1
$$

Pero por submultiplicatividad de la norma:

$$
\|B \mathbf{x}\|_1 \leq \|B\|_1 \|\mathbf{x}\|_1
$$

Entonces:

$$
\|\mathbf{x}\|_1 \leq \|B\|_1 \|\mathbf{x}\|_1
$$

Como $$ \mathbf{x} \neq \mathbf{0} $$ dividimos ambos lados por $$ \|\mathbf{x}\|_1 $$ obteniendo:


$$
1 \leq \|B\|_1
$$

Pero ya probamos que:

$$
\|B\|_1 < 1
$$

lo cual es una contradicción.

Por lo tanto, **( I - B )** debe ser inversible.



---

**Finalmente:**

Como $I - B$ es inversible, $M = \frac{N}{\alpha}(I - B)$ también es inversible. De esta forma, el sistema $M\mathbf{p} = \mathbf{b}$ tiene solución única.



## Punto 3:

```python
def calculaLU(A):
    # matriz es una matriz de NxN
    # Retorna la factorización LU a través de una lista con dos matrices L y U de NxN.
    # Completar! Have fun
    m=A.shape[0]
    n=A.shape[1]
    Ac = A.copy()
    
    if m!=n:
        print('Matriz no cuadrada')
        return
    

    for j in range(n):
        if Ac[j, j] == 0:
            raise ValueError(f"Pivote nulo en columna {j}. No se puede continuar sin pivoteo.")
        for i in range(j + 1, n):
            multiplicador = Ac[i, j] / Ac[j, j]
            Ac[i, j] = multiplicador  # <-- lo guardás acá
            Ac[i, j + 1:] = Ac[i, j + 1:] - multiplicador * Ac[j, j + 1:]
                
            
    L = np.tril(Ac,-1) + np.eye(A.shape[0]) 
    U = np.triu(Ac)
    
    return L, U

def calcula_pagerank(A,alfa):
    # Función para calcular PageRank usando LU
    # A: Matriz de adyacencia
    # d: coeficientes de damping
    # Retorna: Un vector p con los coeficientes de page rank de cada museo
    M = calcular_matriz_M(A, alfa)
    L, U = calculaLU(M) # Calculamos descomposición LU a partir de C y d
    N = A.shape[0]
    b = (N / alfa) * np.ones((N, 1)) # Vector de 1s, multiplicado por el coeficiente correspondiente usando d y N.
    Up = scipy.linalg.solve_triangular(L,b,lower=True) # Primera inversión usando L
    p = scipy.linalg.solve_triangular(U,Up) # Segunda inversión usando U
    return p.flatten()

def graficar_pagerak(A, pageRank, titulo = ""):
    factor_escala = 1e4 # Escalamos los nodos 10 mil veces para que sean bien visibles
    fig, ax = plt.subplots(figsize=(10, 10)) # Visualización de la red en el mapa
    barrios.to_crs("EPSG:22184").boundary.plot(color='gray',ax=ax) # Graficamos Los barrios
    pr = pageRank # Este va a ser su score Page Rank. Ahora lo reemplazamos con un vector al azar
    pr = pr/pr.sum() # Normalizamos para que sume 1
    Nprincipales = 3 # Cantidad de principales
    principales = np.argsort(pr)[-Nprincipales:] # Identificamos a los N principales
    labels = {n: str(n) if i in principales else "" for i, n in enumerate(G.nodes)} # Nombres para esos nodos
    red = nx.from_numpy_array(A) # Construimos la red a partir de la matriz de adyacencia
    # Construimos un layout a partir de las coordenadas geográficas
    red_layout = {i:v for i,v in enumerate(zip(museos.to_crs("EPSG:22184").get_coordinates()['x'],museos.to_crs("EPSG:22184").get_coordinates()['y']))}
    nx.draw_networkx(red,red_layout,node_size = pr*factor_escala, ax=ax,with_labels=False) # Graficamos red
    nx.draw_networkx_labels(red, red_layout, labels=labels, font_size=6, font_color="k") # Agregamos los nombres
    tituloGraph = "Ranking de Museos por PageRank " + titulo + " : \n Museos principales : \n"
    for principal in principales:
        tituloGraph += "Museo " + str(principal) + " : " + str(pr[principal]) + " puntos  \n"
    plt.title(tituloGraph, fontsize=20, pad=20)  # Título grande y separado del gráfico
    plt.axis('off')  # Opcional, para ocultar ejes
    plt.tight_layout()
    plt.show()

m = 3 # Cantidad de links por nodo
mueseos_unidos_3 = construye_adyacencia(D,m)
pageRank = calcula_pagerank(mueseos_unidos_3, 0.2)

graficar_pagerak(mueseos_unidos_3, pageRank, "M = 3, Alfa = 0,2")

```

# Análisis de Sensibilidad del Ranking de Museos

**Comentarios sobre el comportamiento del algoritmo de PageRank bajo variaciones de parámetros:**

---

### 🔹 Influencia del parámetro *m* (vecinos más cercanos)

- Al variar el valor de $m$, se observa una notable varianza en el *top* del PageRank, junto con un desplazamiento geográfico en la ubicación de los museos mejor rankeados.
- Para $m = 1$, los museos 34, 107 y 65 aparecen en el top del ranking. El museo 107 se encuentra en una zona más céntrica, mientras que los otros dos están en áreas periféricas.
- Esto puede explicarse porque, al considerar solo un vecino, es más probable que se visite el museo más cercano, independientemente de su importancia general.
- En este caso, los museos 34 y 65 están muy próximos entre sí, mientras que el 107 está rodeado por otros museos relevantes.
- A medida que $m$ aumenta, la cercanía pierde peso y se empieza a priorizar la importancia del museo. 
- Para $m = 10$, todos los museos rankeados están en zonas céntricas, donde hay una alta densidad de museos muy importantes.

---

### 🔹 Impacto del parámetro *alfa* ($\alpha$)

- Al variar $\alpha$, el ranking cambia poco, aunque se observan diferencias al principio con valores más altos.
- Con un $\alpha$ alto, aumenta la probabilidad de terminar el recorrido, por lo que predominan los museos más conocidos (los que se recorren primero).
- Para $\alpha = 0.5$, aparece en el ranking el museo 93, alejado del centro, lo que sugiere que es muy reconocido y se vuelve una opción atractiva cuando es menos probable que el recorrido termine pronto.
- Cuando $\alpha = \frac{1}{7}$, el ranking cambia completamente hacia museos más periféricos, lo que indica que allí es donde usualmente termina el recorrido de los visitantes.

## Punto 4:



Sabemos que:

-  r = cantidad de museos visitados por cada persona 
-  w = vector que representa el número total de visitantes de cada museo 
- $w_i$ = cantidad de visitantes totales del museo $i$  
-  v = número de personas totales que inician en cada museo
- $v_i$ = número de personas que inician en el museo $i$



$$
\begin{align*}
B = \sum_{k=0}^{r-1} C^k
\end{align*}
$$



$$ C = \text{Matriz de transiciones} $$



---


Entonces,



$$
I \cdot \vec{v} = \text{vector inicial de visitantes } (r = 1)
$$

$$
C \cdot \vec{v} = \text{distribución de visitantes tras un paso } (r = 2)
$$

$$
C^2 \cdot \vec{v} = \text{distribución de visitantes tras dos pasos } (r = 3)
$$

$$
C^{r-1} \cdot \vec{v} = \text{distribución de visitantes tras } r - 1 \text{ pasos}
$$



Por lo tanto, el número de visitantes de \( w \) será la sumatoria de todos estos vectores:

$ \mathbf{w} = C^0 \cdot \mathbf{v} + C^1 \cdot \mathbf{v} + C^2 \cdot \mathbf{v} + C^3 \cdot \mathbf{v} + \cdots + C^{r-1} \cdot \mathbf{v} $ 

Donde I  = $ C^0 $:

Esto puede ser escrito como:

$ \mathbf{w} = \sum_{k=0}^{r-1} C^k \cdot \mathbf{v} $



Con la definición de B tenemos entonces que:

$$
\mathbf{w} = B \cdot \mathbf{v}
$$

Despejamos para encontrar v en trminos de w:

$$
\mathbf{v} = B^{-1} \cdot \mathbf{w}
$$



Asumiendo que B es inversible, queda demostrado que:
$$
v = B^{-1}w 
$$



## Punto 5:

```python
def calcula_matriz_C_continua(D):
    D = D.copy()
    with np.errstate(divide='ignore', invalid='ignore'):
        F = np.where(D != 0, 1 / D, 0)
    np.fill_diagonal(F, 0)
    k = F.sum(axis=1)
    Kinv = np.diag([1 / ki if ki != 0 else 0 for ki in k])
    C = F.T @ Kinv
    return C

def calcula_B(C, cantidad_de_visitas):
    B = np.eye(C.shape[0])  # C^0
    C_potencia = np.eye(C.shape[0])
    for _ in range(1, cantidad_de_visitas):
        C_potencia = np.dot(C_potencia, C)  # C^k
        B += C_potencia
    return B

def calcular_v(w,r):
    C_continua = calcula_matriz_C_continua(D)
    B = calcula_B(C_continua, r)
    L, U = calculaLU(B)
    Up = scipy.linalg.solve_triangular(L,w,lower=True) # Primera inversión usando L
    v = scipy.linalg.solve_triangular(U,Up) # Segunda inversión usando U
    return v

def cargar_visitas(path="visitas.txt"):
    return np.loadtxt(path)

w = cargar_visitas()
r = 3
v = calcular_v(w, r)
C_continua = calcula_matriz_C_continua(D)
B = calcula_B(C_continua, r)

resultados = "Resultados : \n"
for i in range(v.size):
    resultados += f"Museo {i +1} : {v[i]:,.2f} visitantes. \n"

print(resultados)
```

```python
top_n = 10

indices_ordenados = np.argsort(v)[::-1]

# Creamos el string con el top
top_resultados = f"Top {top_n} museos con más visitantes:\n\n"
for i in range(top_n):
    idx = indices_ordenados[i]
    top_resultados += f"Museo {idx + 1} : {v[idx]:,.2f} visitantes\n"

print(top_resultados)
```

## Punto 6:

$$
\mathrm{cond}_{1}(B) = \| B \|_{1} \cdot \| B^{-1} \|_{1}
$$


$$
\frac{\|v - \tilde{v}\|_1}{\|v\|_1} \leq \mathrm{cond}_1(B) \frac{\|w - \tilde{w}\|_1}{\|w\|_1}
$$

$$
\frac{\|w - \tilde{w}\|_1}{\|w\|_1} = 0.05
$$

$$
\frac{\|v - \tilde{v}\|_1}{\|v\|_1} \leq \mathrm{cond}_1(B) \cdot 0.05
$$

```python
norma_B = np.linalg.norm(B, 1)
b = B.sum(axis=1)
Binv = np.diag([1 / bi if bi != 0 else 0 for bi in b])
norma_Binv = np.linalg.norm(Binv, 1)
condicional = norma_B * norma_Binv



def calcular_cota_error_relativo(condicional, porcentaje_error):
    error_relativo_w = porcentaje_error / 100
    cota_error = condicional * error_relativo_w

    return cota_error

cota_error_v = calcular_cota_error_relativo(condicional, 5)
print(f"Cond(B) = {condicional:,.4f}. Cota de error de v = {cota_error_v:,.4f}")
```
